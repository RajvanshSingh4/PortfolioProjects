UNIVARIATE ANALYSIS

```python
categories_col = df[['BR', 'Gender', 'IA', 'Amount of Credit Cards', 'Nationality', 'Occupation Category', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']]

for col in categories_col:
    df[col].value_counts().plot(kind= 'bar')
    plt.title(f"Value Counts for {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

BIVARIATE ANALYSIS

```python
categories_col = df[['BR', 'Gender', 'Amount of Credit Cards', 'Nationality', 'Occupation Category', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']]

for col in categories_col:
    sns.countplot(data = df, x= col, hue = 'Gender')
    plt.title(f"{col} by Gender")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.show()
```

```python
categories_col = df[['BR', 'Gender', 'Amount of Credit Cards', 'Nationality', 'Occupation Category', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']]

for col in categories_col:
    sns.countplot(data = df, x= col, hue = 'Nationality')
    plt.title(f"{col} by Nationality")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.show()
```

HEATMAP

```python
numeric_col = df.select_dtypes(include = float).columns
correlation_matrix = df[numeric_col].corr()

plt.figure(figsize = (12,12))
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.show()
```
